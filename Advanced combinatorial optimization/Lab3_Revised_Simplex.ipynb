# Simplexe révisé

 $$\max  x_1+x_2 + x_3+x_4 $$
$$
\left\{\begin{array}{ccccccccc}
3 x_1   & -& x_2 & + & x_3 & & &  \leq & 4 \\
  & & 2 x_2 & + &   & &x_4 &\leq  & 5 \\
\end{array}\right.$$

 $$ x_1 , x_2,x_3 , x_4  \geq   0 $$

```python
import numpy as np
```

```python

A = np.array([[3, -1, 1 , 0,1, 0], 
              [0, 2.0, 0,1 ,0, 1]])
b = np.array([4,5])
c = np.array([1,1,1,1,0,0])
n=c.shape[0]

def reduced_cost(col) :
    '''OUT : first column from N with reduced cost >0
            otherwise -1
        IN : a set of columns
    '''
    # Base: extraction des colonnes choisies dans la matrice A
    B =  A[:, [ col[x]-1 for x in range(len(col))] ]
    #print("B",B)
    #print("B^-1",np.linalg.inv(B))
    print("xB",np.linalg.inv(B)@b)
    
    colN=[x+1   for x in range(n) if x+1 not in col]
    
    N=A[:,[ colN[x]-1 for x in range(len(colN))]] 
    #print("N",N)
    cB=c[[ col[x]-1 for x in range(len(col))]]
    #print("cB",cB)
    cN=c[[ colN[x]-1 for x in range(len(colN))]] 
    #print("cN",cN)
    # Dans la vraie méthode du SR on calcule les couts individuellement et surtout on ne calcule pas inv(B) (seulement mise à jour)
    rc=cN - cB@(np.linalg.inv(B) @ N)
    print("rc",rc)
    for i in range(len(rc)):
        if rc[i]>0 :
            print("la colonne ", colN[i], "doit entrer en base")
            return(colN[i])
    print("optimum trouvé")
    print("xB",np.linalg.inv(B)@b)
    return(-1)

```

La fonction reduced_cost calcule :
$$r = c -c_B^T B^{-1}N $$

Elle calcule tous les coûts réduits. En pratique, pour simplexe révisé, on ne calcule pas tous les coûts réduits.

```python
# Colonnes de la base (numérotation à partir de 1)
col = [5,6]

i=reduced_cost(col)
B =  A[:, [ col[x]-1 for x in range(len(col))] ]
print("B",B)
print("B^-1.A[:,i]",np.linalg.inv(B) @ A[:,i-1])
print("B^-1.b",np.linalg.inv(B) @ b)
```
